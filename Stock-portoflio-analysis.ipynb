```python
tickers = ['BBCA.JK', 'TLKM.JK', 'ASII.JK', 'UNVR.JK', 'BMRI.JK', '^JKSE']
```

```python
data = yf. download (tickers, start='2020-01-01', end='2023-12-31')
```

```python
data.head ()
```

```python
data.shape
```

```python
data.info ()
```

```python
close_prices = data['Close']
```

```python
close_prices.shape
```

```python
close_prices.head()
```

```python
daily_returns = close_prices.pct_change()
```

```python
daily_returns.head()
```

```python
daily_returns = daily_returns.dropna()
```

```python
daily_returns.head()
```

```python
daily_returns.shape
```

```python
cumulative_returns = (1 + daily_returns).cumprod() - 1
```

```python
cumulative_returns.head()
```

```python
cumulative_returns.tail()
```

```python
plt.figure(figsize=(12,6))
cumulative_returns.plot()
plt.title('Cumulative Return 2021-2023')
plt.xlabel('Tahun')
plt.ylabel('Cumulative Return (%)')
plt.legend()
plt.show()
```

```python
volatility = daily_returns.std()
```

```python
volatility.head()
```

```python
avg_returns = daily_returns.mean()
```

```python
sharpe_ratio = (avg_returns - (0.06/252)) / volatility
```

```python
sharpe_ratio.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(volatility,avg_returns)
for ticker in avg_returns.index :
    plt.annotate(ticker,
                 (volatility[ticker],avg_returns[ticker]))
plt.title('Risk vs Return Analysis')
plt.xlabel('Risk')
plt.ylabel('Return')
plt.show()
```
